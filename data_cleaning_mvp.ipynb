```python
#Dealing duplicate rows

def single_row(df):
    if df.shape[0]==1:    #If there's only one row for that specific player return the data
        return df
    else:
        row = df[df["Team"] == "TOT"]       #If they have multiple teams they payed for just return their total stats (TOT)
        row["Team"] = df.iloc[-1,:]["Team"]   #replace TOT with the last team that player played for in the last row
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
import pandas as pd
import re

award_values = {}

for idx, val in players['Awards'].items():
    if pd.isna(val):
        continue

    # split things like "MVP-10,CPOY-1,AS"
    parts = val.split(',')

    for p in parts:
        match = re.match(r"([A-Za-z0-9]+)-(\d+)", p)
        if match:
            award_name = match.group(1)
            award_value = int(match.group(2))

            if award_name not in award_values:
                award_values[award_name] = {}

            award_values[award_name][idx] = award_value

# create each award column and fill missing values with 0
for award_name, mapping in award_values.items():
    players[award_name] = players.index.map(mapping).fillna(0).astype(int)
```

```python
players['Awards'] = players['Awards'].fillna(0)
```
